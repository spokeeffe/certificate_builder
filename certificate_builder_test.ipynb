```python
# import packages
from reportlab.pdfgen import canvas
from reportlab.lib.utils import ImageReader
import pandas as pd
```

```python
# prototype data
data = {
    'Firstname': 'John',
    'Surname': 'Doe',
    'Email': 'john.doe@example.com',
    'Course': 'Python Programming',
    'Date': '2023-11-03',
    'Instructor': 'Jane Instructor'
}
```

```python
# set some variables
template_path = 'template.pdf'
image_path = 'certificate_image.png'
```

```python
# Read excel document
cert_list_df = pd.read_excel('cert_list_test.xlsx', sheet_name='Sheet1')
cert_list_df['Date'] = cert_list_df['Date'].astype(str)
```

```python
cert_list_df.head()
```

```python
# loop through dataframe to create needed certificates
for index, row in cert_list_df.iterrows():
    output_path = row['Date'] + "_" + row['Course'] + "_" + row['Firstname'] + "_" + row['Surname'] + ".pdf" 
    fill_pdf(image_path, output_path, row)
```

```python
# define builder function

def fill_pdf(image_path, output_path, data):
    c = canvas.Canvas(output_path, pagesize=(1440, 990))
    reader = ImageReader(image_path)
    
    # import the background template image
    c.drawImage(reader, 0,0, 1400, 990)
    
    # concat first and last name
    fullname = data['Firstname'] + " " + data['Surname']
    c.setFont("Times-BoldItalic", 48, leading = None)
    # draw the data
    c.drawString(550, 550, fullname)
    
    c.setFont("Helvetica", 24, leading = None)
    c.drawString(550, 370, data['Course'])
    c.drawString(220, 250, data['Date'])
    c.drawString(800, 250, data['Instructor'])

    c.save()
```

```python
# import packages
from reportlab.pdfgen import canvas
from reportlab.lib.utils import ImageReader
import pandas as pd
```

```python
# prototype data
data = {
    'Firstname': 'John',
    'Surname': 'Doe',
    'Email': 'john.doe@example.com',
    'Course': 'Python Programming',
    'Date': '2023-11-03',
    'Instructor': 'Jane Instructor'
}
```

```python
# set some variables
input_excel = 'inputs/cert_list_test.xlsx'
certificate_map =  'certificate_templates/certificate_map.xlsx'

selected_certificate = 'simon_test'
```

```python
# Read certificate map
cert_map_df = pd.read_excel(certificate_map, index_col=None)
cert_map_df.head()

cert_map_filtered_df = cert_map_df[cert_map_df['template_name']==selected_certificate]
cert_map_filtered_df.head()
image_path = 'certificate_templates/' + cert_map_df['template_image']
print(image_path[0])
image_x_coord = cert_map_filtered_df['image_x'][0]
print(image_x_coord)
```

```python
# Read attendees input
cert_list_df = pd.read_excel(input_excel, sheet_name='Sheet1')
cert_list_df['Date'] = cert_list_df['Date'].astype(str)
```

```python
cert_list_df.head()
```

```python
# loop through dataframe to create needed certificates
for index, row in cert_list_df.iterrows():
    output_path = "certificate_outputs/" + row['Date'] + "_" + row['Course'].replace(" ", "_") + "_" + row['Firstname'].replace(" ", "_") + "_" + row['Surname'].replace(" ", "_") + ".pdf" 
    fill_pdf(selected_certificate, cert_map_filtered_df, output_path, row)
```

```python
# define builder function

def fill_pdf(selected_certificate, cert_map, output_path, data):
    c = canvas.Canvas(output_path, pagesize=(cert_map['image_x'][0], cert_map['image_y'][0]))
    image_path = 'certificate_templates/' + cert_map['template_image'][0]
    reader = ImageReader(image_path)
    
    # import the background template image
    c.drawImage(reader, 0,0, cert_map['image_x'][0], cert_map['image_y'][0])
    
    # concat first and last name
    attendee = data['Firstname'] + " " + data['Surname']
    c.setFont("Times-BoldItalic", 48, leading = None)
    # draw the data
    c.drawCentredString(cert_map['attendee_x'][0], cert_map['attendee_y'][0], attendee)
    
    c.setFont("Helvetica", 24, leading = None)
    c.drawCentredString(cert_map['course_x'][0], cert_map['course_y'][0], data['Course'])
    c.drawCentredString(cert_map['date_x'][0], cert_map['date_y'][0], data['Date'])
    c.drawCentredString(cert_map['instructor_x'][0], cert_map['instructor_y'][0], data['Instructor'])

    c.save()
```
